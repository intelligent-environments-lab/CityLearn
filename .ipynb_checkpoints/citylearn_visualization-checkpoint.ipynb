```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
weather = file[['Outdoor Drybulb Temperature [C]','Outdoor Relative Humidity [%]','Diffuse Solar Radiation [W/m2]','Direct Solar Radiation [W/m2]']]
```

```python
from scipy.ndimage.interpolation import shift
```

```python
n_random = np.random.rand(8760)
```

```python
T_pred_6h = file['Outdoor Drybulb Temperature [C]'] + pd.DataFrame((n_random-.5)*2*.3)[0]
weather['6h Prediction Outdoor Drybulb Temperature [C]'] = pd.DataFrame(shift(T_pred_6h[::-1], 6, cval=0)[::-1])[0]

T_pred_12h = file['Outdoor Drybulb Temperature [C]'] + pd.DataFrame((n_random-.5)*2*.65)[0]
weather['12h Prediction Outdoor Drybulb Temperature [C]'] = pd.DataFrame(shift(T_pred_12h[::-1], 12, cval=0)[::-1])[0]

T_pred_24h = file['Outdoor Drybulb Temperature [C]'] + pd.DataFrame((n_random-.5)*2*1.35)[0]
weather['24h Prediction Outdoor Drybulb Temperature [C]'] = pd.DataFrame(shift(T_pred_24h[::-1], 24, cval=0)[::-1])[0]

RH_pred_6h = file['Outdoor Relative Humidity [%]'] * pd.DataFrame(1-(n_random-.5)*2*.025)[0]
weather['6h Prediction Outdoor Relative Humidity [%]'] = pd.DataFrame(shift(RH_pred_6h[::-1], 6, cval=0)[::-1])[0].clip(0,100)

RH_pred_12h = file['Outdoor Relative Humidity [%]'] * pd.DataFrame(1-(n_random-.5)*2*.05)[0]
weather['12h Prediction Outdoor Relative Humidity [%]'] = pd.DataFrame(shift(RH_pred_12h[::-1], 12, cval=0)[::-1])[0].clip(0,100)

RH_pred_24h = file['Outdoor Relative Humidity [%]'] * pd.DataFrame(1-(n_random-.5)*2*.1)[0]
weather['24h Prediction Outdoor Relative Humidity [%]'] = pd.DataFrame(shift(RH_pred_24h[::-1], 24, cval=0)[::-1])[0].clip(0,100)

Diff_Irradiance_pred_6h = file['Diffuse Solar Radiation [W/m2]'] * pd.DataFrame(1+(n_random-.5)*2*.025)[0]
weather['6h Prediction Diffuse Solar Radiation [W/m2]'] = pd.DataFrame(shift(Diff_Irradiance_pred_6h[::-1], 6, cval=0)[::-1])[0]

Diff_Irradiance_pred_12h = file['Diffuse Solar Radiation [W/m2]'] * pd.DataFrame(1+(n_random-.5)*2*.05)[0]
weather['12h Prediction Diffuse Solar Radiation [W/m2]'] = pd.DataFrame(shift(Diff_Irradiance_pred_12h[::-1], 12, cval=0)[::-1])[0]

Diff_Irradiance_pred_24h = file['Diffuse Solar Radiation [W/m2]'] * pd.DataFrame(1+(n_random-.5)*2*.1)[0]
weather['24h Prediction Diffuse Solar Radiation [W/m2]'] = pd.DataFrame(shift(Diff_Irradiance_pred_24h[::-1], 24, cval=0)[::-1])[0]

Dir_Irradiance_pred_6h = file['Direct Solar Radiation [W/m2]'] * pd.DataFrame(1+(n_random-.5)*2*.025)[0]
weather['6h Prediction Direct Solar Radiation [W/m2]'] = pd.DataFrame(shift(Dir_Irradiance_pred_6h[::-1], 6, cval=0)[::-1])[0]

Dir_Irradiance_pred_12h = file['Direct Solar Radiation [W/m2]'] * pd.DataFrame(1+(n_random-.5)*2*.05)[0]
weather['12h Prediction Direct Solar Radiation [W/m2]'] = pd.DataFrame(shift(Dir_Irradiance_pred_12h[::-1], 12, cval=0)[::-1])[0]

Dir_Irradiance_pred_24h = file['Direct Solar Radiation [W/m2]'] * pd.DataFrame(1+(n_random-.5)*2*.1)[0]
weather['24h Prediction Direct Solar Radiation [W/m2]'] = pd.DataFrame(shift(Dir_Irradiance_pred_24h[::-1], 24, cval=0)[::-1])[0]
```

```python
weather.to_csv('weather_data.csv', index=None, float_format="%.2f")
```
